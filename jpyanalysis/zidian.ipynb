```python
def search_ci_adv(zi, exp, ci):
    se_result = []
    for i in range(len(exp)):
        fit = np.zeros(len(zi))
        for j in range(len(zi)):
            if zi[j] not in exp[i]:
                fit[j] = 1
        if sum(fit)>0:
            continue
        tem_exp = exp[i]
        tem_ci = ci[i]
        data = {
            'ci': tem_ci,
            'explanation': tem_exp,
            '分割线':'------------------------------'
        }
        se_result.append(data)
    return(se_result)
```
